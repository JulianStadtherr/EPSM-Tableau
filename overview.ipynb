```python
# Adjust entries according to assumption
df_wip.loc[20914:20918, 'free_exist'] = 1
df_wip.loc[20914:20918, 'compuls_exempt'] = '2;3;4'
df_wip.loc[20914:20918]
```

```python
# Change format of col compul_exist to int
df_wip['free_exist'] = pd.to_numeric(df_wip['free_exist'], errors='coerce') # Use to_numeric because of nan. Would give error with astype
```

```python
# Remove logical inconsistency: Where there is no free education, it cannot be free by law
df_wip.loc[(df_wip['compuls_exist'] == 1) & (df_wip['free_exist'] == 2), 'free_exist'] = 1
# Check that no rows remain
len(df_wip.loc[(df_wip['compuls_exist'] == 1) & (df_wip['free_exist'] == 2)])
```

```python
# Remove logical inconsistency: Where there is no compulsory ed., no groups can be excluded from it

# Change compuls_ed of Nigeria to some groups excluded because years of comp. ed = 6
df_wip.loc[6741:6744, 'compuls_exist'] = 2
```

```python
# Change remaining cases to no groups excluded
df_wip.loc[(df_wip['compuls_exist'] == 1) & ~((df_wip['compuls_exempt'] == 0) | (df_wip['compuls_exempt'].isna())), 'compuls_exempt'] = 0
# Check that no rows remain
len(df_wip.loc[(df_wip['compuls_exist'] == 1) & ~((df_wip['compuls_exempt'] == 0) | (df_wip['compuls_exempt'].isna()))])
```

```python
# Check for logical inconsistency: where there is supposedly free ed for everyone, yet there are excluded groups
df_wip.loc[(df_wip['compuls_exist'] == 3) & ~((df_wip['compuls_exempt'] == 0) | (df_wip['compuls_exempt'].isna())), 'compuls_exist'] = 2
```

```python
# Whenever there are groups not included in compul. ed. (compul_exist = 2), does that mean there are groups excluded (compuls_exempt != 0 or NA)
x = len(df_wip.loc[(df_wip['compuls_exist'] == 2) & ~((df_wip['compuls_exempt'] == 0) | (df_wip['compuls_exempt'].isna()))])
y = len(df_wip.loc[~((df_wip['compuls_exempt'] == 0) | (df_wip['compuls_exempt'].isna()))])
x/y*100
```

```python
# Check for logical inconsistency: if there are positive years of comp. ed. for countries without comp. ed. 
df1 = df_wip.loc[((df_wip['compuls_exist'] == 1) | (df_wip['compuls_exist'].isna())) & (df_wip['compuls_years'] > 0)]
df1
```

```python
# Change compuls_exist for Urugay & Trinidad to 3
df_wip.loc[[15308,18791], 'compuls_exist'] = 3
```

```python
# Change years of comp. ed. to 0 for remaining countries from df1 (after reloading df1 to remove previously changed rows!)
df1_rows = list(df1.index)
df_wip.loc[df1_rows, 'compuls_years'] = 0
```

```python
# Check which columns have a value for compuls_year between 0-1
df2 = df_wip.loc[df_wip['compuls_years'].between(0, 1, inclusive='neither')]
df2
```
